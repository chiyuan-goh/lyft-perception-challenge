```python
import numpy as np
def process_camvid(label):
    h,w,_ = label.shape
    newlabel = np.zeros((h,w))
#     128 0 192
    newlabel[(label[:,:,0] == 192) & (label[:,:,1] == 0) & (label[:,:,2] == 128)] = 7
    newlabel[(label[:,:,0] == 128) & (label[:,:,1] == 64) & (label[:,:,2] == 128)] = 7
    newlabel[(label[:,:,0] == 128) & (label[:,:,1] == 0) & (label[:,:,2] == 64)] = 10
    newlabel[(label[:,:,0] == 192) & (label[:,:,1] == 128) & (label[:,:,2] == 64)] = 10
    newlabel[(label[:,:,0] == 192) & (label[:,:,1] == 128) & (label[:,:,2] == 192)] = 10
    
    ret = np.zeros_like(label)
    label[:,:,2] = ret
    return label
    
plt.imshow(process_camvid(label))
    
```

```python
from sklearn.utils import shuffle
```

```python
imgs = sorted(os.listdir(imgdir))
labels = sorted(os.listdir(labeldir))
from shutil import copyfile
prefix = "camvid"
limit = 200
imgs, labels = shuffle(imgs, labels)

burl = '/home/cy/Desktop/Train'
dest_img_url = os.path.join(burl, "CameraRGB")
dest_label_url = os.path.join(burl, "CameraSeg")

for i in range(limit):
    copyfile(os.path.join(imgdir, imgs[i]), os.path.join(dest_img_url, "%s_%s" %(prefix, imgs[i]) ) )
    label_img = cv2.imread(os.path.join(labeldir, labels[i]))
    write_img = process_camvid(label_img)
    cv2.imwrite(os.path.join(dest_label_url, "%s_%s"%(prefix, labels[i])), write_img)
    
    
```
